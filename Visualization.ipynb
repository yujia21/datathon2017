```python
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
n = str(1) #chiller number
datatypes = ["Power","ConFlow","EvaFlow","Temp"]
df={}
for datatype in datatypes:
    df[datatype] = pd.read_csv("data/Chiller"+n+"_"+datatype+"_full.csv")
    df[datatype].index=pd.to_datetime(df[datatype]['ts'])
    df[datatype].drop('ts', axis=1, inplace=True)
    print(datatype)
    print(df[datatype].columns)
```

```python
#import datetime
# use .loc[datetime.date(2017,5,1):datetime.date(2017,5,3)] to choose dates

df_clean = df['Power'].drop(df['Power'].ch3Current.argmax())
df_clean[['ch'+str(i+1)+'Watt' for i in range(1)]].plot()
```
